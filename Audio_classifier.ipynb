# Простая модель классифкации на 1D сверточных слоях

```python
class AudioSilero(Dataset):
    def __init__(self,items_frame, labels_frame):
        
        assert type(items_frame) == pd.core.series.Series
        assert type(labels_frame) == pd.core.series.Series
        
        self.items = list(zip(list(items_frame), list(labels_frame)))
        self.length = len(self.items)
        
    def __getitem__(self, index):
        filename, label = self.items[index]
        audioTensor, rate = torchaudio.load(filename)
        return (audioTensor, int(label))     
    
    def __len__(self):
        return self.length
```

```python
device="cpu"
bs=50


train_dataset = AudioSilero(train_files.wav_path, train_files.target)
valid_dataset = AudioSilero(validation_files.wav_path, validation_files.target)
test_dataset = AudioSilero(test_files.wav_path, test_files.target)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size = bs, shuffle = True)
valid_loader = torch.utils.data.DataLoader(valid_dataset, batch_size = bs, shuffle = True)
test_loader  = torch.utils.data.DataLoader(test_dataset,  batch_size = bs, shuffle = True)
```

Каждый wav-файл считывается и записывается в тензор размера 1х48000 (1 канал и 48000 точек сэмплирования)

```python
class SimpleAudioNet(nn.Module): 
    def __init__(self):
        super(SimpleAudioNet, self).__init__() 
        self.features = nn.Sequential(
        nn.Conv1d(1, 16, 80, 4),
        nn.ReLU(),
        nn.MaxPool1d(4),
        nn.Conv1d(16, 32, 3), 
        nn.ReLU(),
        nn.MaxPool1d(4), 
        nn.Conv1d(32, 32, 3), 
        nn.ReLU(),
        nn.MaxPool1d(4), 
        nn.Conv1d(32, 16, 3), 
        nn.ReLU(),
        nn.MaxPool1d(4) 
        )
        self.avgPool = nn.AvgPool1d(30) 
        self.fc1 = nn.Linear(16, 3)
    
    def forward(self, x):
        x = self.features(x)
        x = self.avgPool(x)
        x = x.permute(0, 2, 1) 
        x = self.fc1(x)
        return x.squeeze(1)

```

### Конвейер обучения:

```python
def train(model, optimizer, loss_fn, train_loader, val_loader, epochs=20, device="cpu"):
    for epoch in range(epochs):
        training_loss = 0.0
        valid_loss = 0.0
        model.train()
        for batch in train_loader:
            optimizer.zero_grad()
            inputs, targets = batch
            inputs = inputs.to(device)
#             print(inputs.shape)
            targets = targets.to(device)
            output = model(inputs)
#             print(output.shape)
#             print(targets.shape)
            loss = loss_fn(output, targets)
            loss.backward()
            optimizer.step()
            training_loss += loss.data.item() * inputs.size(0)
#             print('batch Done')
        training_loss /= len(train_loader.dataset)
        
        model.eval()
        num_correct = 0 
        num_examples = 0
        for batch in val_loader:
            inputs, targets = batch
            inputs = inputs.to(device)
            output = model(inputs)
            targets = targets.to(device)
            loss = loss_fn(output,targets) 
            valid_loss += loss.data.item() * inputs.size(0)
            correct = torch.eq(torch.max(F.softmax(output), dim=1)[1], targets).view(-1)
            num_correct += torch.sum(correct).item()
            num_examples += correct.shape[0]
        valid_loss /= len(val_loader.dataset)

        print('Epoch: {}, Training Loss: {:.2f}, Validation Loss: {:.2f}, accuracy = {:.2f}'.format(epoch, training_loss,
        valid_loss, num_correct / num_examples))
```

```python
net_1 = SimpleAudioNet()
net_1.to('cpu')
optimizer = optim.Adam(net_1.parameters(), lr=0.001)
```

```python
train(net_1, optimizer, torch.nn.CrossEntropyLoss(),train_loader, valid_loader, epochs=20)
```

```python
class AudioSilero_MFCC(Dataset):
    def __init__(self,items_frame, labels_frame):
        
        assert type(items_frame) == pd.core.series.Series
        assert type(labels_frame) == pd.core.series.Series
        
        self.items = list(zip(list(items_frame), list(labels_frame)))
        self.length = len(self.items)
        
    def __getitem__(self, index):
        filename, label = self.items[index]
        audioTensor, rate = torchaudio.load(filename)
        MFCC_Tensor = torchaudio.transforms.MFCC(n_mfcc = 70).forward(audioTensor)
        return (MFCC_Tensor, int(label))     
    
    def __len__(self):
        return self.length
```

```python
device="cpu"
bs=50


MFCC_train_dataset = AudioSilero_MFCC(train_files.wav_path, train_files.target)
MFCC_valid_dataset = AudioSilero_MFCC(validation_files.wav_path, validation_files.target)
MFCC_test_dataset = AudioSilero_MFCC(test_files.wav_path, test_files.target)

MFCC_train_loader = torch.utils.data.DataLoader(MFCC_train_dataset, batch_size = bs, shuffle = True)
MFCC_valid_loader = torch.utils.data.DataLoader(MFCC_valid_dataset, batch_size = bs, shuffle = True)
MFCC_test_loader  = torch.utils.data.DataLoader(MFCC_test_dataset,  batch_size = bs, shuffle = True)
```

```python
class MFCC_SimpleAudioNet(nn.Module): 
    def __init__(self):
        super(MFCC_SimpleAudioNet, self).__init__() 
        
        self.features = nn.Sequential(
            
            nn.Conv2d(1, 16, 3),
            nn.ReLU(),
            nn.MaxPool2d(2),
        
            nn.Conv2d(16, 32, 3),
            nn.ReLU(),
            nn.MaxPool2d(2),
    
        )

        self.avgPool = nn.AvgPool2d(2) 
        self.fc1 = nn.Linear(32*8*29, 3)
        

    
    def forward(self, x):
        
        x = self.features(x)
        x = self.avgPool(x)
        x = torch.flatten(x, 1)
        x = self.fc1(x)
        return x.squeeze(1)
```

```python
net = MFCC_SimpleAudioNet()
net.to('cpu')
optimizer = optim.Adam(net.parameters(), lr=0.001)
train(net, optimizer, torch.nn.CrossEntropyLoss(),MFCC_train_loader, MFCC_valid_loader, epochs=20)
```

* Epoch: 3, Training Loss: 0.19, Validation Loss: 0.23, accuracy = 0.91
